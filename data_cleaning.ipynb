Load the data

```python
import pandas as pd

df = pd.read_excel("Crime_2024_with_Weather_Unemp_Monthly.xlsx")
df.head()
```

Drop Unused Columns

```python
df = df.drop(columns=['Crm Cd', 'Crm Cd 2', 'Crm Cd 3', 'Crm Cd 1', 'Crm Cd 4','year', 'date','lat_bin','lon_bin','temp_max','temp_min', 'DR_NO', 'Cross Street', 'Mocodes', 'Year'])
df.head()
```

Rename columns

```python
df.rename(columns={
    'TIME OCC': 'time_of_occurrence',
    'AREA': 'area_code',
    'AREA NAME': 'area_name',
    'Part 1-2': 'crime_part_category',
    'Crm Cd Desc': 'crime_code_description',
    'Vict Age': 'victim_age',
    'Vict Sex': 'victim_sex',
    'Vict Descent': 'victim_descent_code',
    'Premis Cd': 'premises_code',
    'Premis Desc': 'premises_description',
    'Weapon Used Cd': 'weapon_used_code',
    'Weapon Desc': 'weapon_description',
    'Status': 'case_status_code',
    'Status Desc': 'case_status_description',
    'LOCATION': 'location_string',
    'Date Rptd': 'date_reported',
    'DATE OCC': 'date_of_occurrence',
    'Rpt Dist No': 'reporting_district_number',
    'precipitation': 'daily_precipitation_mm',
    'temp_avg': 'daily_avg_temperature_celsius',
    'unemployment_rate': 'unemployment_rate_pct',
    'LAT': 'latitude',
    'LON': 'longitude'
}, inplace=True)

df.head()
```

Victim Descent encoding

```python
descent_mapping = {
    'W': 0,  # White
    'B': 1,  # Black
    'H': 2,  # Hispanic
    'A': 3,  # Asian
    'O': 4,  # Other
    'I': 5,  # American Indian/Alaskan Native
    'P': 6,  # Pacific Islander
    'F': 7,  # Filipino
    'C': 8,  # Chinese
    'J': 9,  # Japanese
    'K': 10, # Korean
    'V': 11, # Vietnamese
    'Z': 12, # Cambodian
    'U': None, # Unknown
    'X': None  # Unknown
}

df['victim_descent_code_new'] = df['victim_descent_code'].map(descent_mapping)
df.head()
```

Victim Sex encoding

```python
sex_mapping = {
    'M': 0,  # Male
    'F': 1,  # Female
    'H': 2, # Other
    'X': None  # Unknown
}

df['victim_sex_new'] = df['victim_sex'].map(sex_mapping)
df.head()
```

Crime Categories

```python
crime_categories = df['crime_code_description'].unique()

for crime in crime_categories:
    print(crime)
```

Set age=0, sex=none, and race=none for stealing Vehicle/Coin/Boat... Crimes

```python
# steal
mask_auto_theft = df['crime_code_description'].str.contains('VEHICLE|COIN|BOAT|DUMPING|TILL', na=False, case=False)

df.loc[mask_auto_theft, 'victim_age'] = 0
df.loc[mask_auto_theft, 'victim_sex'] = None
df.loc[mask_auto_theft, 'victim_race'] = None
df.head()
```

Fill missing victim_age by crime type average

```python
crime_type_avg_age = df.groupby('crime_code_description')['victim_age'].mean()

def fill_age(row):
    if pd.isnull(row['victim_age']):
        return crime_type_avg_age.get(row['crime_code_description'], row['victim_age'])
    else:
        return row['victim_age']

df['victim_age'] = df.apply(fill_age, axis=1)
print(df['victim_age'].isnull().sum())
```

```python
print(df[df['victim_age'].isna()][['crime_code_description']])
```

Fill missing victim_age by overall mean & check none still contains

```python
df['victim_age'] = df['victim_age'].fillna(df['victim_age'].mean())
print(df['victim_age'].isnull().sum())
```

Output the file

```python
df.to_excel("Crime_2024_with_Weather_Unemp_Monthly.xlsx", index=False)
```
